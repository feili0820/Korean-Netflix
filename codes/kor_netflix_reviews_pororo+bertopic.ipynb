## POS

```python
import pandas as pd
df = pd.read_csv('netflix_2017_2021_reviews_KOR.csv')
df = df.where(df.notnull(), None)
df.head(5)
```

```python
titles = df['original_title'].to_list()
reviews = df['original_review'].to_list()
txts= [] 
for x, y in zip(titles, reviews):
  if x == None:
    x = ''
  txt = ' '.join([x,y])
  txts.append(txt.strip())
```

```python
txts[:5]
```

```python
from pororo import Pororo
pos = Pororo(task="pos", lang="ko")
pos_results = []
from tqdm import trange
for x in trange(len(txts)):
  pos_results.append(pos(txts[x]))
```

```python
pos_results[-1]
```

```python
txts[-1]
```

```python
df['pororo_pos'] = pos_results
df.head(5)
```

```python
df.to_csv('netflix_2017_2021_reviews_KOR_pororo.csv', index = None)
```

```python
import pickle
with open('pororo_pos_results.pkl', 'wb') as f:
  pickle.dump(pos_results, f)
```

## Spacing

```python
import pandas as pd
import pickle
with open(r'./reviews_selected.pkl', 'rb') as f:
  df_reviews = pickle.load(f)
df_reviews['google'].head(5)
```

```python
from pororo import Pororo
spacing = Pororo(task="gec", lang="ko")
```

```python
from tqdm import trange
outliers = {}
for x, y in df_reviews.items():
  contents = y['contents'].to_list()
  contents_spacing = []
  outlier = []
  for i in trange(len(contents)):
    txt = contents[i]
    words = txt.split(' ')
    if len(words) > 150:
      outlier.append((i, txt))
      turns = list(range(150, len(words), 150))
      batchs = [spacing(' '.join(words[s:e])) for s, e in zip([0]+turns, turns+[len(words)])]
      txt_spacing = ' '.join(batchs)
    else:
      txt_spacing = spacing(txt)
    contents_spacing.append(txt_spacing)
  y['contents_spacing'] = contents_spacing
  df_reviews[x] = y
  outliers[x]=outlier
  y.head(5)
```

### filtering >=5words

```python
import pickle
with open(r'./reviews_selected_spacingcheck.pkl', 'rb') as f:
  df_reviews = pickle.load(f)
```

```python
df_reviews_filtering = {}
for x, y in df_reviews.items():
  y = y[y['contents_spacing'].apply(lambda x: len(x.split(' '))) > 5]
  filename = ''.join([x, '_netflix_reviews_spacingcheck_5wordsmore.csv'])
  y.to_csv(filename, index=None, encoding='utf-8')
  df_reviews_filtering[x] = y
  print('total number of {0} is {1}.'.format(x, str(len(y))))
```

```python
import pickle
from bertopic import BERTopic
from sentence_transformers import SentenceTransformer
sentence_model = SentenceTransformer("distiluse-base-multilingual-cased-v1")
```

### google

```python
K_G_netflix_model = BERTopic(embedding_model=sentence_model, nr_topics='auto',calculate_probabilities=True,
                top_n_words = 15, low_memory=True, min_topic_size=50)
K_G_netflix_topics, K_G_netflix_probs = K_G_netflix_model.fit_transform(kor_reviews_google)
```

```python
K_G_netflix_model.visualize_topics()
```

```python
K_G_netflix_model.save("K_G_netflix_model")
```

```python
K_G_netflix_fig = K_G_netflix_model.visualize_topics()
K_G_netflix_fig.write_html("K_G_netflix.html")
```

```python
K_G_netflix_model.get_topic_info()
```

```python
K_G_netflix_model.get_topics()
```

```python
K_G_netflix_model.visualize_distribution(K_G_netflix_probs[0])
```

```python
topics_over_time_G = K_G_netflix_model.topics_over_time(kor_reviews_google, K_G_netflix_topics, timestamps_G)
K_G_netflix_model.visualize_topics_over_time(topics_over_time_G)
```

```python
fig_topics_over_time_G = K_G_netflix_model.visualize_topics_over_time(topics_over_time_G)
fig_topics_over_time_G.write_html("K_G_netflix_fig_topics_over_time.html")
```

```python
K_G_netflix_model_paras = {'topics': K_G_netflix_topics , 'probs':K_G_netflix_probs,
              'topics_over_time': topics_over_time_G, 'model': K_G_netflix_model}
with open(r'./K_G_netflix_model_paras_0429.pkl', 'wb') as f:
  pickle.dump(K_G_netflix_model_paras, f)
```
